# 🧾 Missing Values Handling — FuelCast Project

### 1️- Analyse du pourcentage de valeurs manquantes

Description :
Cette étape permet de mesurer la proportion de valeurs manquantes pour chaque variable afin d’identifier les capteurs ou mesures les plus défaillants.

```python
# Calculate missing percentage per column
missing_percent = (
    df_train.isnull().mean() * 100
).sort_values(ascending=False)

# Display top 20 columns with highest missing values
missing_percent.head(20)
```

### 2️ -Visualisation des colonnes les plus impactées

Description :
Une visualisation facilite l’identification rapide des variables nécessitant un traitement spécifique.

```python
# Select top missing columns
top_missing = missing_percent.head(15)

# Plot
plt.figure(figsize=(12,6))

sns.barplot(
    x=top_missing.values,
    y=top_missing.index
)

plt.title("Top Columns with Missing Values")
plt.xlabel("Missing Percentage (%)")
plt.ylabel("Columns")

plt.show()
```

### 3-Segmentation des colonnes selon le taux de Missing

Description :
Les colonnes sont classées en trois catégories pour appliquer des stratégies d’imputation adaptées.

```python
# High missing > 15%
high_missing = missing_percent[
    missing_percent > 15
].index.tolist()

# Medium missing 5–15%
medium_missing = missing_percent[
    (missing_percent > 5) &
    (missing_percent <= 15)
].index.tolist()

# Low missing < 5%
low_missing = missing_percent[
    missing_percent <= 5
].index.tolist()

print("High Missing Columns:", len(high_missing))
print("Medium Missing Columns:", len(medium_missing))
print("Low Missing Columns:", len(low_missing))
```

### 4️- Suppression des colonnes très vides (optionnel)

Description :
Si certaines variables dépassent 40 % de valeurs manquantes, elles peuvent être supprimées car peu exploitables.

```python
threshold = 0.40

df_train = df_train.loc[
    :,
    df_train.isnull().mean() < threshold
]
```

### 5️- Identification des colonnes numériques

Description :
Les méthodes d’imputation statistiques s’appliquent uniquement aux variables numériques.

```python
num_cols = df_train.select_dtypes(
    include="number"
).columns

print("Number of numerical columns:", len(num_cols))
```

### 6️⃣ Imputation simple — Colonnes à faible missing

Méthode : Médiane

Description :
La médiane est robuste face aux valeurs extrêmes fréquentes dans les données mécaniques.

```python
for col in low_missing:
    if col in num_cols:
        df_train[col].fillna(
            df_train[col].median(),
            inplace=True
        )
```

7- Imputation avancée — KNN Imputer

Description :
Les variables moteurs et propulsion étant corrélées, KNN permet une estimation plus réaliste des valeurs manquantes.

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)

# Get numerical columns from the current df_train state for imputation
current_num_cols = df_train.select_dtypes(include="number").columns

df_train[current_num_cols] = imputer.fit_transform(
    df_train[current_num_cols]
)

# After imputation, check for remaining missing values to ensure no NaNs are left
print("Missing values after KNN imputation:", df_train[current_num_cols].isnull().sum().sum())
```

### 8️- Interpolation — Variables de navigation

Description :
Les données de trajectoire du navire évoluent de manière continue, d’où l’utilisation d’une interpolation linéaire.

```python
nav_cols = [
    col for col in df_train.columns
    if "Ship_" in col
]

df_train[nav_cols] = df_train[
    nav_cols
].interpolate(method="linear")
```

### 9️- Interpolation — Variables météorologiques

Description :
Les conditions météo varient progressivement dans le temps, ce qui justifie l’interpolation.

```python
weather_cols = [
    col for col in df_train.columns
    if "Weather_" in col
]

df_train[weather_cols] = df_train[
    weather_cols
].interpolate(method="linear")
```

### - Vérification finale des valeurs manquantes

Description :
Cette étape confirme que toutes les valeurs manquantes ont été correctement traitées.

```python
# Total remaining missing values
total_missing = df_train.isnull().sum().sum()

print("Remaining Missing Values:", total_missing)
```

## Outliers Detection

### 1️- Détection des outliers — Méthode IQR

Description :
La méthode IQR (Interquartile Range) permet d’identifier les valeurs extrêmes situées en dehors de l’intervalle statistique normal.

Formule :

IQR = Q3 − Q1

Outlier < Q1 − 1.5×IQR

Outlier > Q3 + 1.5×IQR

```python
# Select numerical columns
num_cols = df_train.select_dtypes(
    include="number"
).columns

# Detect outliers count per column
outliers_count = {}

for col in num_cols:

    Q1 = df_train[col].quantile(0.25)
    Q3 = df_train[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df_train[
        (df_train[col] < lower_bound) |
        (df_train[col] > upper_bound)
    ]

    outliers_count[col] = outliers.shape[0]

# Display columns with most outliers
pd.Series(outliers_count).sort_values(
    ascending=False
).head(10)
```

### 4-Définition de la fonction Winsorization

Description :
Remplace les valeurs extrêmes par les percentiles 1 % et 99 %.

```python
def winsorize_column(series, lower_q=0.01, upper_q=0.99):

    lower = series.quantile(lower_q)
    upper = series.quantile(upper_q)

    return series.clip(lower, upper)
```

### 5- Application du capping

Description :
On applique la Winsorization à toutes les variables numériques.

```python
for col in num_cols:

    df_train[col] = winsorize_column(
        df_train[col]
    )
```

## Feature Engineering Pipeline

### 1️- Objectif du Feature Engineering

Description :
Le Feature Engineering consiste à créer de nouvelles variables plus explicatives à partir des données brutes afin d’améliorer la performance des modèles de prédiction de consommation fuel.

Dans le contexte FuelCast, nous combinons :

Charges moteurs

Puissance propulsion

Conditions météo

Dynamique du navire

### 2️- Total Engine Fuel Consumption

Description :
On agrège la consommation instantanée de tous les générateurs pour obtenir une mesure globale de la charge énergétique.

```python
engine_fuel_cols = [
    col for col in df_train.columns
    if "MomentaryFuel" in col and "GeneratorEngine" in col
]

df_train["Total_Engine_Fuel"] = df_train[
    engine_fuel_cols
].sum(axis=1)

print("Engine fuel columns used:", engine_fuel_cols)
```

### 3️- Total Shaft Power — Engines

Description :
Représente la puissance mécanique totale fournie par les moteurs.

```python
engine_power_cols = [
    col for col in df_train.columns
    if "ShaftPower" in col and "GeneratorEngine" in col
]

df_train["Total_Engine_ShaftPower"] = df_train[
    engine_power_cols
].sum(axis=1)
```

### 4️ Total Propeller Shaft Power

Description :
Puissance totale fournie aux hélices (propulsion réelle).

```python
propeller_power_cols = [
    "Propeller_Port_ShaftPower",
    "Propeller_Starboard_ShaftPower"
]

df_train["Total_Propeller_Power"] = df_train[
    propeller_power_cols
].sum(axis=1)
```

### 5️- Propulsion Balance Ratio

Description :
Mesure l’équilibre entre hélice bâbord et tribord.

Déséquilibre → surconsommation fuel.

```python
df_train["Propulsion_Balance_Ratio"] = (
    df_train["Propeller_Port_ShaftPower"] /
    (df_train["Propeller_Starboard_ShaftPower"] + 1e-6)
)
```

### 6️-Weather Severity Index

Description :
Indice combinant houle et vent pour mesurer la sévérité des conditions maritimes.

```python
df_train["Weather_Severity_Index"] = (
    df_train["Weather_WaveHeight"] *
    df_train["Weather_WindSpeed10M"]
)
```

### 7️- Ocean Resistance Index

Description :
Mesure la résistance hydrodynamique due aux courants.

```python
df_train["Ocean_Resistance_Index"] = (
    df_train["Weather_OceanCurrentVelocity"] *
    df_train["Ship_SpeedOverGround"]
)
```

### 8️- Effective Propulsion Efficiency

Description :
Compare la puissance propulsion à la vitesse réelle du navire.

```python
df_train["Propulsion_Efficiency"] = (
    df_train["Ship_SpeedOverGround"] /
    (df_train["Total_Propeller_Power"] + 1e-6)
)
```

### 9️- Liste des nouvelles colonnes

```python
new_features = [
    "Total_Engine_Fuel",
    "Total_Engine_ShaftPower",
    "Total_Propeller_Power",
    "Propulsion_Balance_Ratio",
    "Weather_Severity_Index",
    "Ocean_Resistance_Index",
    "Propulsion_Efficiency"
]

df_train[new_features].head()
```

Description :
Les variables issues du Feature Engineering présentent des distributions très asymétriques (skewed), avec des valeurs extrêmes élevées.

La transformation permet de :

Réduire la skewness

Stabiliser la variance

Améliorer la performance des modèles linéaires

### 2- Log Transformation — Balance Ratio

Description :
On applique une transformation logarithmique pour réduire l’asymétrie de la variable représentant l’équilibre de propulsion.

```python
df_train["Propulsion_Balance_Ratio_log"] = np.log1p(
    df_train["Propulsion_Balance_Ratio"]
)
```

### 3- Power Transform — Propulsion Efficiency

Description :
La transformation logarithmique étant insuffisante pour cette variable très asymétrique, on applique une transformation de puissance (Yeo-Johnson).

Cette méthode :

Gère les zéros

Réduit les extrêmes

Normalise la distribution

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer()

df_train["Propulsion_Efficiency_pt"] = pt.fit_transform(
    df_train[["Propulsion_Efficiency"]]
)
```

### Vérification après Power Transform

```python
df_train["Propulsion_Efficiency_pt"] = df_train[
    "Propulsion_Efficiency_pt"
].clip(-3, 3)

```

**Groupe moteurs – Fuel & Power**

Les variables liées aux moteurs, notamment Consumer_GeneratorEngineX_MomentaryFuel, Consumer_GeneratorEngineX_ShaftPower, Consumer_Total_ShaftPower et Total_Engine_Fuel, présentent des corrélations très fortes, comprises entre 0.7 et 0.95. Cette relation est parfaitement cohérente d’un point de vue physique : lorsqu’un moteur fournit davantage de puissance mécanique, il consomme mécaniquement plus de carburant. De plus, les moteurs fonctionnent généralement de manière simultanée, ce qui entraîne une forte colinéarité entre les puissances et consommations individuelles et leurs équivalents agrégés. D’un point de vue data science, conserver les variables globales (Total_Engine_Fuel et Consumer_Total_ShaftPower) est suffisant, tandis que les mesures individuelles par moteur sont redondantes et peuvent être supprimées sans perte d’information.

**Groupe propulsion – Hélices et transmission**

Les variables décrivant la propulsion, telles que Propeller_Port_ShaftPower, Propeller_Starboard_ShaftPower, Propeller_Total_ShaftPower, ainsi que les couples (ShaftTorque) et vitesses de rotation (RotationSpeed), sont fortement corrélées entre elles, avec des coefficients dépassant souvent 0.85. Cette corrélation s’explique par la relation mécanique directe entre couple, vitesse de rotation et puissance, ainsi que par le fonctionnement synchronisé des hélices bâbord et tribord. D’un point de vue analytique, la variable de puissance totale des hélices (Total_Propeller_Power) synthétise efficacement l’information utile. Les variables détaillées par hélice ou par composante mécanique sont donc supprimées afin d’éviter la redondance et la multicolinéarité.

**Navigation – Vitesse du navire**

Les variables Ship_SpeedThroughWater et Ship_SpeedOverGround présentent une corrélation quasi parfaite, proche de 0.9. Cette similarité s’explique par le fait que les courants marins sont généralement faibles ou stables sur la majorité des observations, rendant les deux mesures très proches. ## Préparation des données sans fuite de données

Les variables ont des échelles très différentes (puissance, vitesse, météo), donc le scaling reste nécessaire.
Avant l'entraînement, on corrige surtout un point critique : certaines colonnes contiennent directement de l'information liée à la consommation de fuel. Les garder dans `X` donne un score artificiellement élevé, car le modèle peut reconstruire la cible au lieu de vraiment la prédire.

Amélioration appliquée : on retire de `X` toutes les variables de consommation instantanée (`MomentaryFuel`) ainsi que les variables agrégées dérivées de cette consommation, comme `Total_Engine_Fuel`.


```python
# Identifier les colonnes catégorielles restantes après la réduction des variables
categorical_cols_final = df_train_reduced.select_dtypes(include='object').columns

# Encodage One-Hot des variables catégorielles pour les rendre utilisables par les modèles sklearn
df_train_processed = pd.get_dummies(
    df_train_reduced,
    columns=categorical_cols_final,
    drop_first=True
)

# Variable cible : consommation totale instantanée de fuel
target = "Consumer_Total_MomentaryFuel"

# Correction importante : suppression des variables qui créent une fuite de données.
# Exemple : Total_Engine_Fuel est calculée à partir des consommations fuel des moteurs,
# donc elle donne indirectement la réponse au modèle.
leakage_cols = [
    col for col in df_train_processed.columns
    if ("MomentaryFuel" in col and col != target)
    or col == "Total_Engine_Fuel"
]

# Création de X sans la cible et sans les colonnes de fuite.
# Le modèle devra maintenant prédire la consommation à partir de variables explicatives réelles :
# puissance, vitesse, météo, navigation, propulsion, etc.
X = df_train_processed.drop(columns=[target] + leakage_cols, errors="ignore")
y = df_train_processed[target]

print("Colonnes supprimées pour éviter la fuite de données :")
print(leakage_cols)
print("Nombre de variables explicatives après correction :", X.shape[1])
```

Ajustements Non Linéaires — Comparaison des Corrélations

Description :
Pour choisir le meilleur type d'ajustement, on compare les coefficients de corrélation linéaire après transformation des variables :

- Linéaire    : r(x, y)
- Exponentiel : r(x, ln y)
- Puissance   : r(ln x, ln y)
- Logarithmique : r(ln x, y)

```python
x = df_train_reduced["Ship_SpeedOverGround"].replace(0, np.nan).dropna()
y_target = df_train_reduced.loc[x.index, "Consumer_Total_MomentaryFuel"].replace(0, np.nan).dropna()

# Aligner les deux séries
common_idx = x.index.intersection(y_target.index)
x_clean = x.loc[common_idx]
y_clean = y_target.loc[common_idx]

r_linear  = np.corrcoef(x_clean, y_clean)[0, 1]
r_exp     = np.corrcoef(x_clean, np.log(y_clean))[0, 1]
r_power   = np.corrcoef(np.log(x_clean), np.log(y_clean))[0, 1]
r_log     = np.corrcoef(np.log(x_clean), y_clean)[0, 1]

print("=== Comparaison des ajustements ===")
print(f"Linéaire      r(x, y)         : {r_linear:.4f}")
print(f"Exponentiel   r(x, ln y)      : {r_exp:.4f}")
print(f"Puissance     r(ln x, ln y)   : {r_power:.4f}")
print(f"Logarithmique r(ln x, y)      : {r_log:.4f}")

best = max(
    [("Linéaire", abs(r_linear)), ("Exponentiel", abs(r_exp)),
     ("Puissance", abs(r_power)), ("Logarithmique", abs(r_log))],
    key=lambda t: t[1]
)
print(f"\nMeilleur ajustement : {best[0]} (|r| = {best[1]:.4f})")
```

Visualisation — Courbes de régression polynomiale (degrés 1 à 4)

Comparaison des modèles polynomiaux — R² et RMSE

Description :
On compare des modèles polynomiaux sur les features corrigées. Pour éviter une explosion du nombre de variables, on limite l'expérience aux degrés 1 et 2. Un degré plus élevé sur toutes les features peut créer beaucoup trop d'interactions et favoriser le surapprentissage.


```python
from sklearn.metrics import r2_score, mean_squared_error

results = []

# Après suppression de la fuite de données, on teste seulement les degrés 1 et 2.
# Cela garde le modèle interprétable et limite le risque de surapprentissage.
for degree in [1, 2]:
    poly_pipe = Pipeline([
        ("poly", PolynomialFeatures(degree=degree, include_bias=False)),
        ("scaler", StandardScaler()),
        ("lr", LinearRegression())
    ])
    poly_pipe.fit(X_train, y_train)
    y_pred_poly = poly_pipe.predict(X_test)

    r2   = r2_score(y_test, y_pred_poly)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred_poly))

    results.append({
        "Degré": degree,
        "R²": round(r2, 4),
        "RMSE": round(rmse, 4)
    })
    print(f"Degré {degree} -> R² = {r2:.4f} | RMSE = {rmse:.4f}")

import pandas as pd
results_df = pd.DataFrame(results)
results_df
```

Application du Meilleur Modèle

Description :
D'après la comparaison des métriques, on sélectionne le modèle polynomial au meilleur degré (celui avec le R² le plus élevé sans surapprentissage) et on l'applique comme modèle final.

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# Sélection du meilleur degré parmi les modèles testés sur les features corrigées
best_degree = results_df.loc[results_df["R²"].idxmax(), "Degré"]
print(f"Meilleur degré sélectionné : {best_degree}")

best_poly_pipeline = Pipeline([
    ("poly", PolynomialFeatures(degree=best_degree, include_bias=False)),
    ("scaler", StandardScaler()),
    ("lr", LinearRegression())
])

# Entraînement final du modèle polynomial sans variables de fuite
best_poly_pipeline.fit(X_train, y_train)
y_pred_best = best_poly_pipeline.predict(X_test)

r2_best   = r2_score(y_test, y_pred_best)
rmse_best = np.sqrt(mean_squared_error(y_test, y_pred_best))
mae_best  = mean_absolute_error(y_test, y_pred_best)

print(f"\n=== Modèle Final Corrigé — Polynomial Degré {best_degree} ===")
print(f"R²   : {r2_best:.4f}")
print(f"RMSE : {rmse_best:.4f}")
print(f"MAE  : {mae_best:.4f}")
```

```python
# Tableau récapitulatif des modèles corrigés.
# On évite ici une compréhension de liste trop complexe pour garder le code lisible
# et faciliter l'exécution dans le notebook.
summary_rows = [
    {
        "Modèle": "Régression Linéaire corrigée",
        "R²": round(r2_lr, 4),
        "RMSE": round(rmse_lr, 4),
        "MAE": round(mae_lr, 4)
    }
]

for _, row in results_df.iterrows():
    degree = int(row["Degré"])

    poly_summary_pipeline = Pipeline([
        ("poly", PolynomialFeatures(degree=degree, include_bias=False)),
        ("scaler", StandardScaler()),
        ("lr", LinearRegression())
    ])

    poly_summary_pipeline.fit(X_train, y_train)
    y_pred_summary = poly_summary_pipeline.predict(X_test)

    summary_rows.append({
        "Modèle": f"Polynomial corrigé degré {degree}",
        "R²": round(r2_score(y_test, y_pred_summary), 4),
        "RMSE": round(np.sqrt(mean_squared_error(y_test, y_pred_summary)), 4),
        "MAE": round(mean_absolute_error(y_test, y_pred_summary), 4)
    })

summary = pd.DataFrame(summary_rows)
summary.sort_values("R²", ascending=False)
```

## 1. RandomForest Regressor

Le Random Forest entraîne plusieurs arbres de décision sur des sous-échantillons des données, puis moyenne leurs prédictions. Ce modèle est utile quand les relations entre les variables et la cible ne sont pas strictement linéaires.

Dans cette partie, on entraîne un RandomForestRegressor, on calcule R²/RMSE/MAE, puis on analyse l'importance des variables.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Initialisation du modèle Random Forest.
# n_estimators : nombre d'arbres dans la forêt.
# max_depth limite la profondeur pour réduire le surapprentissage et garder un temps d'exécution raisonnable.
# n_jobs=-1 utilise tous les cœurs disponibles pour accélérer l'entraînement.
rf_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=12,
    min_samples_leaf=5,
    random_state=42,
    n_jobs=-1
)

# Entraînement du modèle sur les données corrigées, sans fuite de données.
rf_model.fit(X_train, y_train)

# Prédiction sur l'ensemble de test.
y_pred_rf = rf_model.predict(X_test)

# Calcul des métriques d'évaluation.
r2_rf = r2_score(y_test, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
mae_rf = mean_absolute_error(y_test, y_pred_rf)

print("=== RandomForestRegressor — Test Set ===")
print(f"R²   : {r2_rf:.4f}")
print(f"RMSE : {rmse_rf:.4f}")
print(f"MAE  : {mae_rf:.4f}")
```

Si une variable inattendue apparaît en tête, il faut vérifier son sens métier et s'assurer qu'elle ne contient pas indirectement une information trop proche de la cible.

## 3. Gradient Boosted Trees Regressor

Le Gradient Boosting entraîne les arbres de manière séquentielle. Chaque nouvel arbre cherche à corriger les erreurs des arbres précédents. Ce modèle est souvent performant sur des relations non linéaires, mais il est plus sensible aux hyperparamètres que Random Forest.

```python
from sklearn.ensemble import GradientBoostingRegressor

# Initialisation du modèle Gradient Boosting.
# learning_rate contrôle la contribution de chaque arbre.
# max_depth limite la complexité de chaque arbre faible.
gbt_model = GradientBoostingRegressor(
    n_estimators=150,
    learning_rate=0.05,
    max_depth=3,
    min_samples_leaf=5,
    random_state=42
)

# Entraînement du modèle sur les données corrigées.
gbt_model.fit(X_train, y_train)

# Prédiction sur l'ensemble de test.
y_pred_gbt = gbt_model.predict(X_test)

# Calcul des métriques d'évaluation.
r2_gbt = r2_score(y_test, y_pred_gbt)
rmse_gbt = np.sqrt(mean_squared_error(y_test, y_pred_gbt))
mae_gbt = mean_absolute_error(y_test, y_pred_gbt)

print("=== GradientBoostingRegressor — Test Set ===")
print(f"R²   : {r2_gbt:.4f}")
print(f"RMSE : {rmse_gbt:.4f}")
print(f"MAE  : {mae_gbt:.4f}")
```

## 4. Comparaison avec les modèles précédents

On compare maintenant les modèles de la séance 5 avec la régression linéaire corrigée et le meilleur modèle polynomial corrigé. Les métriques utilisées sont :

- R² : qualité globale de l'ajustement
- RMSE : erreur quadratique moyenne, sensible aux grandes erreurs
- MAE : erreur absolue moyenne, plus facile à interpréter

```python
# Tableau comparatif final incluant les modèles de la séance 5.
# Les variables r2_lr, rmse_lr, mae_lr, r2_best, rmse_best et mae_best viennent des sections précédentes.
models_comparison = pd.DataFrame([
    {
        "Modèle": "Régression Linéaire corrigée",
        "R²": r2_lr,
        "RMSE": rmse_lr,
        "MAE": mae_lr
    },
    {
        "Modèle": f"Polynomial corrigé degré {best_degree}",
        "R²": r2_best,
        "RMSE": rmse_best,
        "MAE": mae_best
    },
    {
        "Modèle": "RandomForestRegressor",
        "R²": r2_rf,
        "RMSE": rmse_rf,
        "MAE": mae_rf
    },
    {
        "Modèle": "GradientBoostingRegressor",
        "R²": r2_gbt,
        "RMSE": rmse_gbt,
        "MAE": mae_gbt
    }
])

models_comparison = models_comparison.sort_values("R²", ascending=False).reset_index(drop=True)
models_comparison[["R²", "RMSE", "MAE"]] = models_comparison[["R²", "RMSE", "MAE"]].round(4)
models_comparison
```

```python
# Visualisation comparative des métriques.
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

metric_info = [
    ("R²", "Plus élevé = meilleur"),
    ("RMSE", "Plus faible = meilleur"),
    ("MAE", "Plus faible = meilleur")
]

for ax, (metric, subtitle) in zip(axes, metric_info):
    sns.barplot(
        data=models_comparison,
        x=metric,
        y="Modèle",
        hue="Modèle",
        palette="Set2",
        legend=False,
        ax=ax
    )
    ax.set_title(f"Comparaison {metric} - {subtitle}")
    ax.set_xlabel(metric)
    ax.set_ylabel("")

plt.tight_layout()
plt.show()
```

## 5. Visualisation des prédictions

On trace les prédictions en fonction des vraies valeurs. Plus les points sont proches de la diagonale rouge, plus le modèle prédit correctement la consommation.

```python
# Pour garder le graphique lisible, on affiche un échantillon du test set.
plot_sample = min(3000, len(y_test))
plot_index = y_test.sample(n=plot_sample, random_state=42).index

prediction_plots = [
    ("Régression linéaire", y_pred_lr),
    ("RandomForest", y_pred_rf),
    ("Gradient Boosting", y_pred_gbt)
]

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharex=True, sharey=True)

for ax, (model_name, predictions) in zip(axes, prediction_plots):
    sampled_true = y_test.loc[plot_index]
    sampled_pred = pd.Series(predictions, index=y_test.index).loc[plot_index]

    ax.scatter(sampled_true, sampled_pred, alpha=0.25, s=12)
    min_val = min(sampled_true.min(), sampled_pred.min())
    max_val = max(sampled_true.max(), sampled_pred.max())
    ax.plot([min_val, max_val], [min_val, max_val], color="red", linewidth=2)
    ax.set_title(model_name)
    ax.set_xlabel("Valeurs réelles")
    ax.set_ylabel("Prédictions")

plt.suptitle("Prédictions vs valeurs réelles")
plt.tight_layout()
plt.show()
```

## 6. Bonus — Impact de quelques hyperparamètres

On teste rapidement plusieurs configurations pour observer l'impact du nombre d'arbres et de la profondeur maximale sur RandomForest. Cette expérimentation permet de voir si l'amélioration de performance justifie l'augmentation de complexité.

```python
# Petit test d'hyperparamètres RandomForest.
# Pour limiter le temps d'exécution, on teste seulement quelques configurations
# et on utilise un sous-échantillon du train set.
rf_param_results = []

sample_size = min(20000, len(X_train))
X_train_tuning = X_train.sample(n=sample_size, random_state=42)
y_train_tuning = y_train.loc[X_train_tuning.index]

for n_estimators in [50, 100]:
    for max_depth in [8, 12, None]:
        rf_candidate = RandomForestRegressor(
            n_estimators=n_estimators,
            max_depth=max_depth,
            min_samples_leaf=5,
            random_state=42,
            n_jobs=-1
        )
        rf_candidate.fit(X_train_tuning, y_train_tuning)
        y_pred_candidate = rf_candidate.predict(X_test)

        rf_param_results.append({
            "n_estimators": n_estimators,
            "max_depth": "None" if max_depth is None else max_depth,
            "R²": r2_score(y_test, y_pred_candidate),
            "RMSE": np.sqrt(mean_squared_error(y_test, y_pred_candidate)),
            "MAE": mean_absolute_error(y_test, y_pred_candidate)
        })

rf_param_results_df = pd.DataFrame(rf_param_results)
rf_param_results_df[["R²", "RMSE", "MAE"]] = rf_param_results_df[["R²", "RMSE", "MAE"]].round(4)
rf_param_results_df.sort_values("R²", ascending=False)
```

```python
# Visualisation de l'impact des hyperparamètres RandomForest sur le RMSE.
plt.figure(figsize=(9, 5))
sns.barplot(
    data=rf_param_results_df,
    x="max_depth",
    y="RMSE",
    hue="n_estimators",
    palette="Set1"
)
plt.title("Impact des hyperparamètres RandomForest sur le RMSE")
plt.xlabel("Profondeur maximale")
plt.ylabel("RMSE")
plt.tight_layout()
plt.show()
```
